```python
Cancer = pd.read_csv('BreastCancer.csv')
```

```python
Cancer
```

```python
Cancer.drop(['Id'], axis = 1 , inplace = True )
```

```python
Cancer.replace('0', -999999)
```

```python
def RetBin(X):
    if X == "M":
        return 0
    else:
        return 1
Cancer['Diagnosis'] = Cancer['Diagnosis'].map(RetBin)  
# Cancer['Diagnosis'] = Cancer['Diagnosis'].map(lambda x : 1 if x == "M" else 0)
```

```python
Cancer
```

```python
import numpy as np 
X = np.array(Cancer.drop(["Diagnosis"], axis = 1))
y = np.array(Cancer['Diagnosis'])
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

#Training and testing the models
[X_train, X_test, y_train, y_test] = train_test_split(X,y, test_size = 0.1,random_state = 0)
# SVC Classifiers

Classifiers = SVC()

model = Classifiers.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print("Accuracy of SVC: ", accuracy)
```

```python
#logistic Regression 
from sklearn.linear_model import LogisticRegression

Classifiers = LogisticRegression(solver = 'liblinear')
model = Classifiers.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print("Accuracy of Logistic Regression : ", accuracy)
```

```python
# Prediction 

import pickle
pickle.dump(model, open("LogisticRegression.m", "wb"))

# Loading the model

loaded_model = pickle.load(open("LogisticRegression.m", "rb"))
accuracy = loaded_model.score(X_test, y_test)
print("Accuracy of Logistic Regression : ", accuracy)
```
